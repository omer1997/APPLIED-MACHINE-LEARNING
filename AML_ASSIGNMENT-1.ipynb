1. Pick a graph from http://www.tylervigen.com/spurious-correlations and recreate it using matplotlib (you can also use numpy and pandas). You can generate the data manually by looking at the points. Ensure the axes are labeled properly.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as tick

fig, ax1 = plt.subplots(figsize=(20, 5))
plt.title(label="Age of Miss America\n correlates with\n Murders by steam, hot vapours and hot objects\n correlation:87.01%(r = 0.870127)\n\n\n", fontsize=20, color="black") 
plt.grid(axis="y")  # if you dont want the grid lines in the graph comment this and try 

#data
years =[1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009]
red_data =[24, 24, 24, 21, 22, 21, 24, 22, 20, 19, 22]
black_data =[7.3, 7.3, 7.3, 3, 4, 3, 8, 4, 2, 3, 2]

x=np.arange(18.75, 26, 1.25)    # y axis data starting from 18.75 till 25 with 1.25 step over

ax1.set_ylim(18, 25)   # how much data you want to view on the y axis 18-25 range
ax1.set_ylabel("Age of Miss America", fontsize=18, color="red")
ax1.set_xticks(years)    # year on the x axis 1999-2009
ax1.set_yticks(x)     # 18.75-25 on y axis

red_line, =ax1.plot(years, red_data, color="red", lw=1, marker='D', markersize=6) # red line year vs age data , color of line , line width, square box

for label in ax1.get_yticklabels():   # making the color of y axis data ( 18.75- 25 ) red
    label.set_color("red")

def year_format(x, y):          # required for making y axis data like 18 years .... years after the data
    return '{} yrs'.format(x)  # [data 18.75 - 25]
ax1.yaxis.set_major_formatter(tick.FuncFormatter(year_format))

red_line.set_label('Age of Miss America') # text of legend box
ax1.legend(bbox_to_anchor=(0.6, -0.3)) # space between boxes, height position 
 

y= np.arange(2, 10, 2)  # twinx data [2 - 8 murders]
ax2 = ax1.twinx()      # used to create axis opposite to y axis
ax2.set_yticks(y)    # data on axis
ax2.set_ylabel("Murders by Stream", fontsize=18, color="black") # Label on twinx 
black_line, =ax2.plot(years, black_data, lw=1, color="black", marker='o', markersize=6)  # plotting black line

for label in ax2.get_yticklabels():     
    label.set_color("black") 

def murder_format(x1, y):
    return '{} murders'.format(x1)
ax2.yaxis.set_major_formatter(tick.FuncFormatter(murder_format))

black_line.set_label('Murders by Stream')
ax2.legend(bbox_to_anchor=(0.4, -0.3))

# top section of graph axis    
secondary_axis= ax1.secondary_xaxis('top')  # adds another x axis on top
secondary_axis.set_xticks(years)             # ticks for years
for label in secondary_axis.get_xticklabels():   # coloring the year red
    label.set_color("red")
```

3. Reproduce the graphs on overlapping data from Fundamentals of Data Visualization https://serialmentor.com/dataviz/overlapping-points.html that is figures 18.1 to 18.4, using matplotlib as subplots in a single figure. You can read the data from following link. Please use the weblink for reading the data, not the local path.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/pratoolbharti/NIU/master/Applied_Machine_Learning_S20/Homework-1/mpg.csv")
```

```python
data
```

```python
fwd_data = data.loc[data['drv'] == 'f']
fwd_data
```

```python
rwd_data = data.loc[data['drv'] == 'r']
```

```python
rwd_data
```

```python
fourwd_data = data.loc[data['drv'] == '4']
```

```python
fourwd_data
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1
fig, ax = plt.subplots(2, 2, figsize=(16, 16))
line1 = ax[0, 0].scatter(fwd_data['displ'], fwd_data['cty'],color="orange",s=80)
line2 = ax[0, 0].scatter(rwd_data['displ'], rwd_data['cty'],color="blue",s=80)
line3 = ax[0, 0].scatter(fourwd_data['displ'], fourwd_data['cty'],color="black",s=80)
ax[0,0].legend((line1, line2,line3),("FWD", "RWD","4WD"),title='drive train')
ax[0,0].set_xlabel('displacement (l)')
ax[0,0].set_ylabel('fuel economy (mpg)')
ax[0,0].set_title("City fuel economy versus engine displacement (bad)")
#ax[0,0].set_title("Figure 18.1")


#plot 2
# If we make individual points partially transparent, then overplotted points appear as darker points
#  and thus the shade of the points reflects the density of points in that location of the graph

line4 = ax[0, 1].scatter(fwd_data['displ'], fwd_data['cty'],color="orange",alpha=0.4,s=80)
line5 = ax[0, 1].scatter(rwd_data['displ'], rwd_data['cty'],color="blue",alpha=0.4,s=80)
line6 = ax[0, 1].scatter(fourwd_data['displ'], fourwd_data['cty'],color="black",alpha=0.4,s=80)
ax[0,1].legend((line4, line5,line6),("FWD", "RWD","4WD"),title='drive train')
ax[0,1].set_xlabel('displacement (l)')
ax[0,1].set_ylabel('fuel economy (mpg)')
#ax[0,1].set_title("Figure 18.2")
ax[0,1].set_title("City fuel economy versus engine displacement with transparency")




#plot 3
# apply a small amount of jitter to the points, i.e., 
# to displace each point randomly by a small amount in either the x or the y direction or both.

height_jitter = fwd_data['displ'] + np.random.normal(0, 0.3, size=len(fwd_data))
weight_jitter = fwd_data['cty'] + np.random.normal(0, 0.3, size=len(fwd_data))

height_jitter1 = rwd_data['displ'] + np.random.normal(0, 0.3, size=len(rwd_data))
weight_jitter1 = rwd_data['cty'] + np.random.normal(0, 0.3, size=len(rwd_data))

height_jitter2 = fourwd_data['displ'] + np.random.normal(0, 0.3, size=len(fourwd_data))
weight_jitter2 = fourwd_data['cty'] + np.random.normal(0, 0.3, size=len(fourwd_data))

line7 = ax[1, 0].scatter(height_jitter,weight_jitter,color="orange",alpha=0.4,s=80)
line8 = ax[1, 0].scatter(height_jitter1,weight_jitter1,color="blue",alpha=0.4,s=80)
line9 = ax[1, 0].scatter(height_jitter2,weight_jitter2,color="black",alpha=0.4,s=80)
ax[1,0].legend((line7, line8,line9),("FWD", "RWD","4WD"),title='drive train')
ax[1,0].set_xlabel('displacement (l)')
ax[1,0].set_ylabel('fuel economy (mpg)')
#ax[1,0].set_title("Figure 18.3")
ax[1,0].set_title("City fuel economy versus engine displacement \nwith transparency with small jitter")


#plot 4
# If we jitter too much, we end up placing points in locations that are not representative of the underlying dataset. 
#The result is a misleading visualization of the data.

height_jitter3 = fwd_data['displ'] + np.random.normal(0, 0.7, size=len(fwd_data))
weight_jitter3 = fwd_data['cty'] + np.random.normal(0, 0.7, size=len(fwd_data))

height_jitter4 = rwd_data['displ'] + np.random.normal(0, 0.7, size=len(rwd_data))
weight_jitter4 = rwd_data['cty'] + np.random.normal(0, 0.7, size=len(rwd_data))

height_jitter5 = fourwd_data['displ'] + np.random.normal(0, 0.7, size=len(fourwd_data))
weight_jitter5 = fourwd_data['cty'] + np.random.normal(0, 0.7, size=len(fourwd_data))

line10 = ax[1, 1].scatter(height_jitter3,weight_jitter3,color="orange",alpha=0.4,s=80)
line11 = ax[1, 1].scatter(height_jitter4,weight_jitter4,color="blue",alpha=0.4,s=80)
line12 = ax[1, 1].scatter(height_jitter5,weight_jitter5,color="black",alpha=0.4,s=80)

ax[1,1].legend((line10, line11,line12),("FWD", "RWD","4WD"),title='drive train')
ax[1,1].set_xlabel('displacement (l)')
ax[1,1].set_ylabel('fuel economy (mpg)')
#ax[1,1].set_title("Figure 18.4")
ax[1,1].set_title("City fuel economy versus engine displacement \nwith transparency with more jitter (bad)")
```
